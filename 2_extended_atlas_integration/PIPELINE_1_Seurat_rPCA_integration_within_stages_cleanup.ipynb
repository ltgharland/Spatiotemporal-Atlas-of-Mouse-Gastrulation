## rPCA Integration within Time Points

```r
all_genes <- rownames(atlas_ss_downsample)
```

```r
#provide seurat object to integrate
seurat_object <- atlas_ss_downsample

### rPCA within STAGES ###
# Load the parallel package
library(parallel)

# Set parameters
num_features <- 2000
k_anchor <- 5
k_weight <- 100
features_to_integrate <- all_genes

# Define the stages you want to integrate
stages_to_integrate <- c(
                    #"Mixed gastrulation", 
                    #"E6.5_Original", 
                    #"E6.75_Original", 
                    #"E7.0_Original", 
                    #"E7.25_Original", 
                    #"E7.5_Original", 
                    #"E7.75_Original",
                    #"E8.0_Original",
                    #"E8.25_Original",
                    "E8.5_Original",
                    "E8.5_Extension"
                    #"E8.75_Extension",
                    #"E9.0_Extension",
                    #"E9.25_Extension",
                    #"E9.5_Extension"
    )

# Create a list to store integrated objects
integrated_objects <- list()

# Set the number of cores to use (adjust as needed)
num_cores <- 20

# Parallelize the processing of stages
integrated_objects <- mclapply(stages_to_integrate, function(stage) {
    message(stage)
  # Subset the data for the current stage
  object <- seurat_object[, seurat_object@meta.data$stage_updated == stage]
  
  # Split object based on embryo
  object.split <- SplitObject(object, split.by = "sample")
  
  # Normalize and identify variable features for each dataset independently
  object.split <- lapply(X = object.split, FUN = function(x) {
    x <- FindVariableFeatures(x, selection.method = "vst", nfeatures = num_features)
  })
  
  features <- SelectIntegrationFeatures(object.list = object.split, nfeatures = num_features)
  
  # Select features that are repeatedly variable across datasets for integration
  object.split <- lapply(X = object.split, FUN = function(x) {
    x <- ScaleData(x, features = features, verbose = FALSE)
    x <- RunPCA(x, features = features, verbose = FALSE)
  })
  
  # Find Integration Anchors
  if (length(unique(object@meta.data$sample)) > 1) {   
    rPCA_anchors <- FindIntegrationAnchors(object.list = object.split, 
                                           anchor.features = features, 
                                           reduction = "rpca", 
                                           k.anchor = k_anchor,
                                           scale = FALSE
    )

    # Integrate the data
    object_integrated <- IntegrateData(anchorset = rPCA_anchors,
                                       k.weight = k_weight,
                                       features.to.integrate = all_genes,
                                       new.assay.name = "within_timepoint_integration"
                                       ) 
    # Return the integrated object
    return(object_integrated)
  } else {
    # Return the unmodified object for single-sample cases
    return(object.split[[1]])
  }
}, mc.cores = 20)

# Now, integrated_objects contains the integrated data for each stage
names(integrated_objects) <- stages_to_integrate
```

```r
### WRITE A FUNCTION TO: extract VariableFeatures from a list of Seurat objects

extract_variable_features <- function(object_list) {
  variable_features_list <- lapply(object_list, function(seurat_object) {
    VariableFeatures(seurat_object)
  })
  return(variable_features_list)
}

## USE THE FUNCTION: Assuming integrated_objects is your list of Seurat objects
all_variable_features <- extract_variable_features(integrated_objects)

# Combine all VariableFeatures into a single vector and make it unique
combined_variable_features <- unique(unlist(all_variable_features))
str(combined_variable_features)

## USE THE OVERLAPPING IDENTIFIED VARIABLE FEATURES TO PERFORM PCA ON ALL THE DATATSETS ACROSS TIMEPOINTS
overlapping_features <- combined_variable_features

# Define the dimensions and cluster resolution to use
dims_to_use <- 1:30  # Adjust as needed
cluster_resolution <- 2  # Adjust as needed

# Number of CPU cores to use for parallel processing
num_cores <- 20  # Adjust this as needed

# Create a cluster for parallel processing
cl <- makeCluster(num_cores)

# Define a function to perform the specified operations on a Seurat object
perform_operations <- function(seurat_object) {
  # Scale and run PCA
  seurat_object <- ScaleData(seurat_object,
                            features = overlapping_features)
  seurat_object <- RunPCA(seurat_object,
                         features = overlapping_features)
  
  # Elbow plot
  ElbowPlot(seurat_object, ndims = 50)
  
  # Run UMAP, Find Neighbors, and Find Clusters
  seurat_object <- RunUMAP(seurat_object, dims = dims_to_use) %>%
    FindNeighbors(., reduction = "pca", dims = dims_to_use) %>%
    FindClusters(., resolution = cluster_resolution)
  
  return(seurat_object)
}

# Apply the specified operations to each Seurat object in parallel using mclapply
processed_objects <- mclapply(integrated_objects, perform_operations, mc.cores = num_cores)

# Close the cluster
stopCluster(cl)
```

## Integration within an Atlas Version (Original and Extended)

```r
# Subset of stages for integration across a subset of timepoints
subset_stage_list_1 <- list(processed_objects$E6.5_Original,
                            processed_objects$E6.75_Original,
                            processed_objects$E7.0_Original,
                            processed_objects$E7.25_Original,
                            processed_objects$E7.5_Original,
                            processed_objects$E7.75_Original,
                            processed_objects$E8.0_Original,
                            processed_objects$E8.25_Original,
                            processed_objects$E8.5_Original
                           )
  
subset_stage_list_2 <- list(processed_objects$E8.5_Extension,
                            processed_objects$E8.75_Extension,
                            processed_objects$E9.0_Extension,
                            processed_objects$E9.25_Extension,
                            processed_objects$E9.5_Extension
                           )

# Function to perform integration for a given subset of stages
perform_integration <- function(object_list, k_anchor, k_weight) {
  anchors <- FindIntegrationAnchors(object.list = object_list, 
                                     anchor.features = overlapping_features, 
                                     reduction = "rpca", 
                                     k.anchor = k_anchor,
                                     scale = FALSE)
  object_integrated <- IntegrateData(anchorset = anchors, 
                                     k.weight = k_weight,
                                     new.assay.name = "within_atlas_version_integration")
  return(object_integrated)
}

# Set parameters
k_anchor <- 5
k_weight <- 200

# List of subsets of stages
subset_stage_lists <- list(subset_stage_list_1, subset_stage_list_2)

# Initialize a list to store integrated results for each subset
integrated_results <- list()

# Perform integration in parallel for each subset of stages
integrated_results <- mclapply(subset_stage_lists, function(object_list) {
  perform_integration(object_list, k_anchor, k_weight)
}, mc.cores = parallel::detectCores())

# Store the integrated results in separate variables (if needed)
integrated_subset_stage_list_1 <- integrated_results[[1]]
integrated_subset_stage_list_2 <- integrated_results[[2]]
```

```r
# Apply the specified operations to each Seurat object in parallel using mclapply
processed_objects_round_2 <- mclapply(integrated_results, perform_operations, mc.cores = num_cores)

# Give names to the objects in the processed_objects_round_2 list
processed_objects_round_2
names(processed_objects_round_2) <- c("Original", "Extensions")
```
